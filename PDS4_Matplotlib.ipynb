<h2> 2.  Plotting curves and adding a legend </h2>

<br>

```python
# Different ways in which a plot can be customized:

x = np.linspace(0, 10, 200)
sine_x = np.sin(x)

# Set a plot size and dpi
plt.figure(figsize=(11,6), dpi=200)

# Plot f(x) with a 'green' colour, set the line width, and label the line as 'sine'
[line_handle] = plt.plot(x, sine_x, c='m', linewidth=4, label='sine')

# Plot a second function of x
cosine_x = np.cos(x)

# #RRGGBB
[line_handle2] = plt.plot(x, cosine_x, c='r', linewidth=2, label='cos')

# Create a legend for the plot
plt.legend(handles=[line_handle, line_handle2], fontsize=30, loc='lower right', shadow=True)

# Customize the axis ticks, axis labels, and the plot title
plt.xticks(fontsize=20)
plt.yticks([-1, -0.5, 0, 0.5, 1], [-1, -0.5, 0, 0.5, 1], fontsize=20)
plt.xlabel('x axis', fontsize=30)
plt.ylabel('y axis', fontsize=30)
plt.title('Lets have a plot', fontsize=30)
plt.show()
```

<h2> 3. Drawing a customized scatter plot </h2>

<br>

```python
# A customized scatter plot

# Generate random data from a 2-dimensional Gaussian distribution
x1 = np.random.normal(loc=[0,0], scale=1, size=(50,2))
print('x1 :\n', x1, '\n')

plt.figure(figsize=(9,6), dpi=150)

# Create a scatter plot marked with 'o' points
g1 = plt.scatter(
    x1[:,0], x1[:,1], marker='o', s=80, c='w', edgecolor='g',
    label='Gaussian #1'
)

# Generate more random data from 2-dimensional Gaussian distributions
x2 = np.random.normal(loc=[10,-1], scale=0.5, size=(50,2))
x3 = np.random.normal(loc=[8,5], scale=1.5, size=(50,2))


# Plot x2 and x3
g2 = plt.scatter(x2[:,0], x2[:,1], marker='x', s=80, c='r',  label='Gaussian #2'
)
g3 = plt.scatter(
    x3[:,0], x3[:,1], marker='d', s=80, c='w', edgecolor='b',
    label='Gaussian #3'
)

# Position the legend
plt.legend(
        handles=[g1, g2, g3], fontsize=30, shadow=True, 
    bbox_to_anchor=(1.03, 1), borderaxespad=0
)

plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.xlabel('Feature 1', fontsize=30)
plt.ylabel('Feature 2', fontsize=30)
plt.title('Mixture of 3 Gaussians', fontsize=30)
plt.show()
```

<h2> 4. Visualization data with 3D Plot </h2>

<br>

```python
# Plotting in 3D

import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure(dpi=120)
ax = plt.axes(projection='3d')

n = 100
xs = np.random.rand(n)
ys = np.random.rand(n)
zs = np.random.rand(n)



# Plot points with z < 0.5 marked differently than z >= 0.5
mask = zs < 0.5
c = ['r', 'b']
m = ['o', '^']
ax.scatter(xs[mask==0], ys[mask==0], zs[mask==0], c=c[0], marker=m[0])
ax.scatter(xs[mask==1], ys[mask==1], zs[mask==1], c=c[1], marker=m[1])

#Labelling the axis
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')


plt.title("3D Plotting")

#Viewing the 3-D plot from different angles
ax.view_init(45, 45)

plt.show()
```
